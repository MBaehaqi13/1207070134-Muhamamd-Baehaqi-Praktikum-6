# Flip Gambar

Import dependency

```python
import numpy as np # mengimpor pustaka NumPy dan mengaitkannya dengan alias 'np'
import imageio # mengimpor pustaka ImageIO, yang menyediakan fungsi untuk membaca dan menulis berbagai format file gambar
import matplotlib.pyplot as plt # mengimpor modul pyplot dari pustaka Matplotlib dan mengaitkannya dengan alias 'plt'
```

Membaca gambar

```python
img = imageio.imread("beatles.jpg") 
```

Mendapatkan resolusi dan type dari gambar

```python
img_height = img.shape[0] # Mengambil ketinggian
img_width = img.shape[1] # Mengambil lebar 
img_channel = img.shape[2] # Mengambil jumlah channel 
img_type = img.dtype # Mengambil tipe data gambar 
```

Membuat variabel dengan resolusi dan tipe yang sama seperti gambar

```python
img_flip_horizontal = np.zeros(img.shape, img_type) 
img_flip_vertical = np.zeros(img.shape, img_type) 
```

Membalik gambar secara horizontal

```python
for y in range(0, img_height): # Looping vertikal
    for x in range(0, img_width):  # Looping horizontal
        for c in range(0, img_channel): # Looping channel warna
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c] # Melakukan flipping horizontal pada gambar
```

Membalik gambar secara vertical

```python
for y in range(0, img_height): # Looping vertikal
    for x in range(0, img_width): # Looping horizontal
        for c in range(0, img_channel): # Looping channel warna
            img_flip_vertical[y][x][c] = img[img_height-1-y][x][c] # Melakukan flipping vertical pada gambar
```

Menampilkan hasil balik gambar

```python
plt.imshow(img_flip_horizontal) # Menampilkan gambar 
plt.title("Flip Horizontal") # Memberi judul 
plt.show() # Menampilkan gambar 
plt.imshow(img_flip_vertical) # Menampilkan gambar 
plt.title("Flip Vertical") # Memberi judul 
plt.show() # Menampilkan gambar 
```
