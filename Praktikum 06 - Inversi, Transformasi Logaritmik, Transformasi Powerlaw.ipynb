Import Library

```python
import numpy as np # mengimpor pustaka NumPy dan mengaitkannya dengan alias 'np'
import imageio # mengimpor pustaka ImageIO, yang menyediakan fungsi untuk membaca dan menulis berbagai format file gambar
import matplotlib.pyplot as plt # mengimpor modul pyplot dari pustaka Matplotlib dan mengaitkannya dengan alias 'plt'
```

Membaca GambarÂ¶

```python
img = imageio.imread("beatles.jpg")
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
```

# Inversi

Membuat variabel img_inversi

```python
img_inversi = np.zeros(img.shape, dtype=np.uint8)
```

Membuat fungsi untuk inversi grayscale

```python
def inversi_grayscale(nilai):
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = nilai - gray
            img_inversi[y][x] = (gray, gray, gray)
```

Membuat fungsi untuk inversi rgb

```python
def inversi_rgb(nilai):
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            red = nilai - red
            green = img[y][x][1]
            green = nilai - green
            blue = img[y][x][2]
            blue = nilai - blue
            img_inversi[y][x] = (red, green, blue)
```

Menampilkan hasil inversi

```python
inversi_grayscale(255)
plt.imshow(img_inversi)
plt.title("Inversi Grayscale")
plt.show()

inversi_rgb(255)
plt.imshow(img_inversi)
plt.title("Inversi RGB")
plt.show()
```

# Log

Membuat variabel img_log untuk menampung hasil

```python
img_log = np.zeros(img.shape, dtype=np.uint8)
```

Mendefinisikan fungsi untuk log

```python
def log(c):
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = int(c * np.log(gray + 1))
            if gray > 255:
                gray = 255
            if gray < 0:
                gray = 0
            img_log[y][x] = (gray, gray, gray)
```

Menampilkan hasil log

```python
log(30)
plt.imshow(img_log)
plt.title("Log")
plt.show()
```

# Inversi & Log

Membuat variabel img_inlog untuk menampung hasil

```python
img_inlog = np.zeros(img.shape, dtype=np.uint8)
```

Mendefinisikan fungsi untuk inversi log

```python
def inlog(c):
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = int(c * np.log(255 - gray + 1))
            if gray > 255:
                gray = 255
            if gray < 0:
                gray = 0
            img_inlog[y][x] = (gray, gray, gray)
```

Menampilkan hasil inversi log

```python
inlog(30)
plt.imshow(img_inlog)
plt.title("Inversi & Log")
plt.show()
```

# Nth Power

Membuat variabel img_nthpower untuk menampung hasil

```python
img_nthpower = np.zeros(img.shape, dtype=np.uint8)
```

Mendefinisikan fungsi untuk nth power

```python
def nthpower(c, y):
    thc = c / 100
    thy = y / 100
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = int(thc * pow(gray, thy))
            if gray > 255:
                gray = 255
            if gray < 0:
                gray = 0
            img_nthpower[y][x] = (gray, gray, gray)
```

Menampilkan hasil

```python
nthpower(50, 100)
plt.imshow(img_nthpower) # pernyataan yang menggunakan fungsi imshow dari modul pyplot yang diimpor sebelumnya dengan alias plt
plt.title("Nth Power") # pernyataan yang menggunakan fungsi title dari modul pyplot untuk memberikan judul pada plot
plt.show() # enampilkan plot yang telah dibuat
```

# Nth Root Power

Membuat variabel img_nthrootpower

```python
img_nthrootpower = np.zeros(img.shape, dtype=np.uint8)
```

Membuat fungsi untuk nth root power

```python
def nthrootpower(c, y):
    thc = c / 100
    thy = y / 100
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = int(thc * pow(gray, 1./thy))
            if gray > 255:
                gray = 255
            if gray < 0:
                gray = 0
            img_nthpower[y][x] = (gray, gray, gray)
```

Menampilkan hasil

```python
nthrootpower(50, 100)
plt.imshow(img_nthrootpower)
plt.title("Nth Root Power")
plt.show()
```
